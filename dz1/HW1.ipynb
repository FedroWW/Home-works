```python
# Номер3

a = str(input())
word={'E': '3', 'J': 'L', 'S': '2', 'Z': '5', 'L': 'J', '2': 'S', '5': 'Z', '3': 'E', 'A': 'A', 'H': 'H', 'I': 'I', 'M': 'M', 'O': 'O', 'T': 'T', 'U': 'U', 'V': 'V', 'W': 'W', 'X': 'X', 'Y': 'Y', '1': '1', '8': '8'}
same = 'AHIMOTUVWXY18'

zerk = True #потом будет удобно вызвать как элементы и их отображения- это элементы и зеркальной строки и зеркального палиндрома
same = True #элементы зеркальной строки

for i in range(len(a) // 2):
    if a[i] != a[-1 * (i + 1)]: #проверка симметричных элементов на зеркальность троки
        same = False
    if a[i] != word.get(a[-1 * (i + 1)]): #проверка симметричных элементов как на зеркальность строки так и палиндрома
        zerk = False


if word and len(a) % 2 != 0: #проверка на "нерегулярность"
    if a[len(a) // 2] not in same:
           zerk = False

if zerk and same:
    print("{p} is a mirrored palindrome.".format(p=a))
elif zerk:
    print("{p} is a mirrored string.".format(p=a))
elif same:
    print("{p} is a regular palindrome.".format(p=a))
else:
    print("{p} is not a palindrome.".format(p=a))
```
